## Types of Skewness and Kurtosis


## Skewness
Skewness measures the asymmetry of a data distribution around its mean. It helps to understand if the data leans more toward one side of the distribution.

### Types of Skewness
- **Positive Skew (Right-Skewed)**:
  - The tail on the right side is longer.
  - Most data points are concentrated on the left.
  - Mean > Median > Mode.
  - **Example**: Income distributions where most people earn less, but a few earn very high amounts.

- **Negative Skew (Left-Skewed)**:
  - The tail on the left side is longer.
  - Most data points are concentrated on the right.
  - Mean < Median < Mode.
  - **Example**: Age of retirement, where most people retire at similar ages but a few retire much earlier.

- **Symmetrical Skew (No Skew)**:
  - The distribution is symmetrical.
  - Mean = Median = Mode.
  - **Example**: A perfect normal distribution.



```python

# Visualizing types of skewness
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate datasets
symmetrical_data = np.random.normal(0, 1, 1000)
positive_skew_data = np.random.exponential(scale=1, size=1000)
negative_skew_data = -np.random.exponential(scale=1, size=1000)

# Plotting
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(symmetrical_data, kde=True, color='blue')
plt.title('Symmetrical Skew (No Skew)')

plt.subplot(1, 3, 2)
sns.histplot(positive_skew_data, kde=True, color='green')
plt.title('Positive Skew (Right-Skewed)')

plt.subplot(1, 3, 3)
sns.histplot(negative_skew_data, kde=True, color='red')
plt.title('Negative Skew (Left-Skewed)')

plt.tight_layout()
plt.show()
    
```

## Kurtosis
Kurtosis measures the "tailedness" or the presence of outliers in a distribution. It shows how much of the data lies in the tails compared to the peak.

### Types of Kurtosis
- **Mesokurtic (Normal Kurtosis)**:
  - Similar to a normal distribution.
  - Indicates a moderate number of outliers.
  - **Example**: Standard normal distribution.

- **Leptokurtic (Heavy-Tailed)**:
  - Higher peak and heavier tails.
  - Indicates many outliers.
  - **Example**: Financial returns where extreme values occur more frequently.

- **Platykurtic (Light-Tailed)**:
  - Flatter peak and lighter tails.
  - Indicates fewer outliers.
  - **Example**: Uniform distribution.


```python

# Visualizing types of kurtosis
mesokurtic_data = np.random.normal(0, 1, 1000)
leptokurtic_data = np.concatenate([np.random.normal(0, 1, 800), np.random.normal(0, 0.1, 200)])
platykurtic_data = np.random.uniform(-2, 2, 1000)

# Plotting
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(mesokurtic_data, kde=True, color='blue')
plt.title('Mesokurtic (Normal Kurtosis)')

plt.subplot(1, 3, 2)
sns.histplot(leptokurtic_data, kde=True, color='green')
plt.title('Leptokurtic (Heavy-Tailed)')

plt.subplot(1, 3, 3)
sns.histplot(platykurtic_data, kde=True, color='red')
plt.title('Platykurtic (Light-Tailed)')

plt.tight_layout()
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
data = np.random.normal(50, 10, 1000)  # Normal distribution with mean=50, std=10

# Calculate Interquartile Range (IQR)
q1 = np.percentile(data, 25)  # First quartile
q3 = np.percentile(data, 75)  # Third quartile
iqr = q3 - q1  # Interquartile range
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")

# Plot Histogram
plt.figure(figsize=(10, 5))
sns.histplot(data, kde=True, color='blue')
plt.title('Histogram of Data Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Plot Boxplot
plt.figure(figsize=(5, 5))
sns.boxplot(data=data, color='orange')
plt.title('Boxplot with Interquartile Range')
plt.ylabel('Values')
plt.show()
```

# Hypothesis Testing

##  Z-Test

The **Z-Test** is used to test the mean of a population when the standard deviation is known and the sample size is large (typically n > 30).

### Formula
For a one-sample Z-test:

$$ Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}} $$

Where:
- $ \bar{X} = sample mean $
- $ \mu = population mean $
- $ \sigma = population standard deviation $
- $ n\ = sample size $



```python
from statsmodels.stats.weightstats import ztest

# Sample data
sample_data = [51, 45, 33, 45, 67, 34, 56, 44, 56]
population_mean = 50

# Perform Z-Test
z_stat, p_value = ztest(sample_data, value=population_mean)
print(f"Z-statistic: {z_stat}, P-value: {p_value}")
```

## T-Test

The **T-Test** is used to compare the means of two groups when the population standard deviation is unknown and the sample size is small (n < 30). It is commonly used for hypothesis testing in situations where data is assumed to follow a normal distribution.

### Types of T-Tests:
1. **One-Sample T-Test**: Compares the sample mean to a known value (typically the population mean).
2. **Independent Two-Sample T-Test**: Compares the means of two independent groups.
3. **Paired Sample T-Test**: Compares the means of two related groups (e.g., before and after measurements).

### Formula
For a **one-sample T-test**:
$$
t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}
$$
Where:
- $ \bar{X} = sample mean $
- $ \mu = population mean $
- $ s = sample standard deviation $
- $ n = sample size $$

The **Two-Sample T-Test** is used to compare the means of two independent groups to determine if there is a significant difference between them. This test assumes that the two groups are independent and that the data in each group follows a normal distribution.

### Formula

For the **Two-Sample T-Test**, the test statistic is calculated as:

$$
t = \frac{\bar{X_1} - \bar{X_2}}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

Where:
- $ \bar{X_1} $ and $ \bar{X_2} $ are the sample means of the two groups
- $ s_1 $ and $ s_2 $ are the sample standard deviations of the two groups
- $ n_1 $ and $ n_2 $ are the sample sizes of the two groups



```python
from scipy import stats

# Sample data
group1 = [20, 21, 19, 22, 20]
group2 = [30, 29, 31, 32, 30]

# Perform Two-Sample T-Test
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## ANOVA (Analysis of Variance)

**ANOVA** is used to compare the means of three or more groups to determine if at least one of the group means is significantly different from the others.

### Types of ANOVA:
1. **One-Way ANOVA**: Compares the means of three or more independent groups based on one factor.
2. **Two-Way ANOVA**: Compares the means of groups based on two factors and evaluates the interaction between those factors.

### Formula
For a **One-Way ANOVA**, the F-statistic is calculated as:
$$
F = \frac{MS_{between}}{MS_{within}}
$$
Where:
- $ MS_{between} $ = Mean Square Between Groups
- $MS_{within} $ = Mean Square Within Groups


```python
from scipy.stats import f_oneway

# Sample groups
group1 = [20, 21, 19, 22, 20]
group2 = [30, 29, 31, 32, 30]
group3 = [40, 39, 41, 38, 40]

# Perform ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

## Chi-Squared Test

The **Chi-Squared Test** is used to determine if there is a significant association between two categorical variables or to test if a sample comes from a population with a specific distribution. There are two main types of Chi-Squared Tests:
1. **Chi-Squared Goodness-of-Fit Test**: Determines if a sample comes from a population with a specific distribution.
2. **Chi-Squared Test of Independence**: Determines if two categorical variables are independent.

### Formula
For the **Chi-Squared Goodness-of-Fit Test**, the Chi-Squared statistic is calculated as:

$$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

Where:
- $ O_i  = observed frequency $
- $ E_i  = expected frequency $
 

```python
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [50, 30, 20]
expected = [40, 30, 30]

# Perform Chi-Squared Test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
print(f"Chi2 Statistic: {chi2_stat}, P-value: {p_value}")
```


### **Summary of Tests:**

| **Test**        | **Purpose**                                             | **Formula**                                              |
|-----------------|---------------------------------------------------------|----------------------------------------------------------|
| **Z-Test**      | Compare sample mean with population mean (large sample) |  Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}  |
| **T-Test**      | Compare means of two groups (small sample)              | \( t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}} \)       |
| **ANOVA**       | Compare means across three or more groups               | \( F = \frac{MS_{between}}{MS_{within}} \)               |
| **Chi-Squared** | Test association between categorical variables          | \( \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} \)            |


### Sampling
- Sampling involves selecting a subset of data from a larger dataset (population) to infer insights about the population.
- Types of sampling:
  - Random Sampling
  - Stratified Sampling
  - Cluster Sampling
  - Systematic Sampling

### Central Limit Theorem (CLT)
- The CLT states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution.
- Key point: Larger sample sizes yield better approximations.

### Law of Large Numbers (LLN)
- The LLN states that as the number of trials increases, the sample mean will converge to the population mean.
- This principle ensures the reliability of statistics from large samples.

### Five-Number Summary
- A descriptive statistic that includes:
  1. Minimum
  2. First Quartile (Q1)
  3. Median (Q2)
  4. Third Quartile (Q3)
  5. Maximum
    

```python

# Five-Number Summary
import numpy as np

# Sample data
data = [7, 15, 36, 39, 40, 41, 42, 43, 46, 49]

# Calculate the five-number summary
min_val = np.min(data)
q1 = np.percentile(data, 25)
median = np.median(data)
q3 = np.percentile(data, 75)
max_val = np.max(data)

print(f"Five-Number Summary:")
print(f"Minimum: {min_val}")
print(f"Q1: {q1}")
print(f"Median: {median}")
print(f"Q3: {q3}")
print(f"Maximum: {max_val}")
    
```
